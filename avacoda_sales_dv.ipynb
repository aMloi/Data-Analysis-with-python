# Which avocado size is most popular?

Avocados are increasingly popular and delicious in guacamole and on toast. The Hass Avocado Board keeps track of avocado supply and demand across the USA, including the sales of three different sizes of avocado.using line plot to figure out which size is the most popular.

## creating an avocado dataframe

```python
import pandas as pd
import numpy as np

# Create random data for demonstration
np.random.seed(0)
dates = pd.date_range(start='2015-01-01', end='2018-12-31', periods=1014)
types = np.random.choice(['conventional', 'organic'], size=1014)
years = np.random.randint(2015, 2019, size=1014)
avg_prices = np.random.uniform(0.5, 2.5, size=1014)
sizes = np.random.choice(['small', 'medium', 'large', 'extra_large'], size=1014)
nb_sold = np.random.randint(1000, 10000000, size=1014)

# Create DataFrame
avocados = pd.DataFrame({
    'date': dates,
    'type': types,
    'year': years,
    'avg_price': avg_prices,
    'size': sizes,
    'nb_sold': nb_sold
})

# Display DataFrame shape
print(avocados.shape)
```

```python
import matplotlib.pyplot as plt

# Looking at the first few rows of data
print(avocados.head())

# Getting the total number of avocados sold of each size
nb_sold_by_size = avocados.groupby('size')['nb_sold'].sum()
print(nb_sold_by_size)
```

```python
# making a bar plot of the number of avocados sold by size
nb_sold_by_size.plot(kind='bar',title='the number of avocados sold by size')

# Show the plot
plt.show()
```

# visualize the change in avocado sales over three years.

```python
avocados['year'] = pd.to_datetime(avocados['date']).dt.year

# Getting the total number of avocados sold for each year
nb_sold_by_year = avocados.groupby('year')['nb_sold'].sum()

# Create a line plot of the number of avocados sold by year
nb_sold_by_year.plot(kind='line', xlabel='Year', ylabel='Number sold', title='Number of Avocados Sold by Year')

# Show the plot
plt.show()
```

```python
# Scatter plot of avg_price vs. nb_sold with title
avocados.plot(y ='avg_price', x= 'nb_sold', title = '"Number of avocados sold vs. average price"',kind = 'scatter')
plt.show()
```

##create multiple histograms to compare the prices of conventional and organic avocados

```python
# Histogram of conventional avg_price
avocados[avocados["type"]== "conventional"]['avg_price'].hist(alpha=0.9,bins=20)

# Histogram of organic avg_price
avocados[avocados["type"]== "organic"]['avg_price'].hist(alpha=0.4,bins=20)
# Add a legend
plt.legend(['conventional','organic'])

# Show the plot
plt.show()
```
